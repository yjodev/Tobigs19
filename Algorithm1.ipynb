# 문제 1. 승리 카드
카드들이 N × M 형태로 놓여 있을 때, 게임의 룰에 맞게 승리한 카드의 숫자를 출력하는 프로그램
을 작성하시오.
[게임 룰]
1. N명의 투빅이들은 각각 M개의 숫자 카드를 골라 차례대로 한 줄 씩 일렬로 놓을 수 있다. 최
종적으로 N × M 형태로 숫자카드들이 놓여진다. 여기서 N은 행의 개수를 의미하고, M은 열
의 개수를 의미한다.
2. 먼저, 자신이 놓은 카드 M개 중 가장 작은 숫자 카드를 뽑는다.
3. 그 다음, N명의 투빅이들이 고른 가장 작은 숫자 카드 N개 중 가장 높은 숫자를 가진 투빅이
가 승리한다.
[입력 조건]
- 첫째 줄에 숫자 카드들이 놓인 행의 개수 N과 열의 개수 M이 공백을 기준으로 하여 각각 자연
수로 주어진다.
- 둘째 줄부터 N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어진다. 각 숫자는 1 이상 10,000 이하
의 자연수이다.
[출력 조건 ]
- 첫째 줄에 게임의 룰에 맞게 승리한 카드에 적힌 숫자를 출력한다.

```python
# 행과 열 입력받기
n, m = map(int, input().split())
```

```python
result = 0
# 앞서 입력받은 열의 개수만큼 숫자를 입력받는다.
for i in range(n): # 행의 개수만큼 반복
    cards = list(map(int, input().split())) # 입력받은 값 list화하여 분리
    card_min = min(cards)     # 현재 입력받은 행에서 최소값 찾기
    result = max(result, card_min)     # 각 행의 최소값 중에서 최대값 찾기
    
print(result)
```

# 문제2
맹구가 키우는 달팽이가 N × N 크기의 정사각형 미로 위에 서 있다. 이 공간은 1 곱하기
1 크기의 정사각형으로 나누어져 있으며 가장 왼쪽 위 좌표는 (1,1)이고, 가장 오른쪽 아
래 좌표는 (N,N)이다. 맹구는 달팽이가 상, 하, 좌, 우 방향으로 스스로 이동할 수 있도록
달팽이를 훈련시키고자 한다. 달팽이의 시작 좌표는 항상 (1,1)이다. 맹구가 미로에 R, L,
U, D 중의 문자를 반복적으로 적어 달팽이에게 보여주면 달팽이는 미로에 적힌 문자를
보고 차례대로 이동을 한다. 달팽이가 도착하는 최종 위치가 정답 위치라면 맹구는 달팽
이에게 깻잎을 선물로 준다. 이때 달팽이가 N × N 크기의 정사각형 공간을 벗어나는 움
직임은 무시된다. 그때, 주어진 입력에 따라 도착한 달팽이의 도착 지점의 좌표를 출력하
세요.

다음은 N=5인 지도와 그림판 내용이다.
이 경우 6개의 명령을 받아 달팽이가 움직이는 위치는 순서대로 (1,2), (1,3), (1,3), (2,3),
(3,3), (3,2) 이므로 최종적으로 달팽이가 도착하게 되는 곳의 좌표는 (3,2)이다. 다시 말해
3행 2열의 위치에 해당하므로 (3,2)라고 적는다.
[입력 조건]
- 첫째 줄에 공간의 크기를 나타내는 N이 주어진다. (1≤N≤100)
- 둘째 줄에 달팽이가 이동해야 하는 방향이 적힌 미로 내용이 주어진다. (1≤이동 횟수≤100)

[출력 조건]
- 첫째 줄에 달팽이가 최종적으로 도착해야 하는 지점의 좌표 (X, Y)를 공백으로 구분하여 출력한다.

[입력 예시]
5
R R U D D L

[출력 예시]
3 2



```python
# 좌표 공간의 크기 입력
N = int(input())
```

```python
# 초기 좌표 설정
x, y = 1,1

# 경로 입력받기
directions = input().split()


for d in directions:
    if (d == 'U'):
        nx = x + -1 # x 좌표는 -1
        ny = y + 0  # y 좌표는 그대로
    elif (d == 'D'):
        nx = x + 1
        ny = y + 0
    elif (d == 'L'):
        nx = x + 0
        ny = y + -1
    else:
        nx = x + 0
        ny = y + 1

    # 좌표계를 벗어날 경우    
    if (nx >= 1 and nx <= 5 and ny >= 1 and ny <= 5):
        x = nx
        y = ny

# 도착 지점의 좌표 출력
print(x, y)
```

# 문제 3. 가보자고
소영이의 대학원 건물 엘리베이터는 한 번 작동 시 U층만큼 올라갔다가, D층만큼 내려간다.
이 엘리베이터를 타고 건물 최상층인 H층까지 올라가려고 한다.
건물은 0층부터 H층까지 로 구성되어 있으며, H층에 도착하면 엘리베이터가 내려가지 않는다. 소영이가 이
엘리베이터를 타고 H층에 도착하려면, 몇 번 동작 시켜야 할까요?
[입력 조건]
- 첫째 줄에 세 정수 U, D, H가 공백으로 구분되어 주어진다.
- 1 <= D <= U <= H <= 1000000000

[출력 조건]
- 첫째 줄에 H층에 도착하기까지 엘리베이터를 작동시켜야 하는 횟수를 출력한다.
[입력 예시1] [출력 예시1]
2 1 5 4

[입력 예시2] [출력 예시2]
5 1 6 2

```python
# 엘리베이터 한 번 작동시 U-D 만큼 이동하며, 총 H만큼 이동해야 한다.
# 마지막 날에는 U만큼 이동한다.
# U*n - B*(n-1) = V 인 n을 찾아야 한다.
# n >= (V-D)/(U-D)

import math
U, D, H = map(int, input().split()) # U,D,H 입력받기

n = math.ceil((H-D)/(U-D))

print(n)
```

# 문제 4. 숫자 바꾸기

재민이는 0과 1로 이루어진 카드 번호 C를 가지고 있다. 카드 번호 C의 숫자를 전부 같게 만들어야 하는데, 재민이
가 할 수 있는 것은 카드 번호 C에서 연속된 하나 이상의 숫자를 잡고 모두 바꾸는 것이다. 바꾸는 것은 1을 0으로, 0
을 1로 뒤집는 것을 의미한다.
예를 들어 문자열 C가 00011000일 때,

1. 전체를 바꾸면 11100111이 된다.
2. 4번째 문자부터 5번째 문자까지 바꾸면 1111111이 되어서 2번 만에 모두 같은 숫자로 만들 수 있다.
하지만, 처음부터 4번째 문자부터 5번째 문자를 바꾸면 한 번에 0000000이 되어서 1번 만에 같은 문자로 뒤집을
수 있다. 문자열이 주어졌을 때, 재민이가 해야 하는 행동의 최소 횟수를 출력 하시오.


[입력 조건]
첫째 줄에 카드 번호 C가 주어진다. 카드 번호 C의 길이는 100만보다 작다.
[출력 조건]
첫째 줄에 재민이가 해야 하는 행동의 최소 숫자를 출력한다.

[입력 예시]
11111

[출력 예시]
0

```python
C = input()
 
arrange_to_0 = 0 # 문자열 모두 0으로 바꾸는 경우
arrange_to_1 = 0 # 문자열 모두 1로 바꾸는 경우
 
if C[0] == '1': # 첫번째 원소가 1인 경우
  arrange_to_0+=1
else: # 첫번째 원소가 0인 경우
  arrange_to_1+=1
 
for i in range(1,len(C)-1): # 두번째부터 마지막까지 반복
  if C[i] != C[i+1]: # 현재 숫자와 뒷 숫자가 같지 않을 때 
    # 뒷 숫자를 현재수와 같게 바꾼다.
    # 뒷 숫자를 0으로 바뀌는 경우
    if C[i+1] == '1':
      arrange_to_0+=1
    # 뒷 숫자를 1로 바뀌는 경우
    else:
      arrange_to_1+=1
 

print(min(arrange_to_0,arrange_to_1))
```

# 문제 5. 우영우수
영우는 자신의 이름과 같이 거꾸로 읽어도, 앞으로 읽어도 같은 수를 우영우수라고 부른다.
예를 들어 121, 12421을 우영우수라고 할 수 있다. 123, 1231은 뒤에서부터 읽으면 다르므로 우영우수가 아
니다. 이 문제에서는 010과 같이 앞에 무의미한 0이 오는 경우가 없다고 하자.
[입력 조건]
입력은 여러 개의 테스트 케이스로 이루어져 있으며, 각 줄마다 0보다 크고 100000보다 작은 수가 주어진다.
입력 마지막 줄에는 0이 주어지며, 이 줄은 문제에 포함되지 않는다.
[출력 조건]
각 줄마다 주어진 수가 우영우수면 ‘yes’, 아니면 ‘no’를 출력한다.
[입력 예시]
817
707
0

[출력 예시]
no
yes

```python
num = 1
while True:
  num = input() # str로 입력받기
  if num == '0': # 0을 입력받는다면 반복을 멈추고 빠져나온다
	  break

  if num == num[::-1]: # 앞뒤가 같은지 확인
      print("yes")
  else:
      print("no")
```
