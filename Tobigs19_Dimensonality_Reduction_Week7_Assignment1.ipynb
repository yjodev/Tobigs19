### 1-5) ë§ˆì§€ë§‰ìœ¼ë¡œ ê³ ìœ  ë²¡í„° ì¶•ìœ¼ë¡œ ê°’ì„ ë³€í™˜í•©ë‹ˆë‹¤

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]): # range(3)
        if i == 0:
            new = [np.dot(X,eigenvectors[:,i])] 
            # Xì™€ eigenvectorì„ ë‚´ì í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ ìƒì„±
            # print( i ,"new:",new)
        else: 
            new = np.concatenate((new, np.dot(X, eigenvectors.T[i]).reshape(1,-1)),axis=0) # reshapeë¥¼ ì‚¬ìš©í•´ ë³´ì•˜ë‹¤.    
            # concatencateë¥¼ ì´ìš©í•˜ì—¬ new list ì— Xì™€ i(1,2)ë¥¼ ë‚´ì í•œ ê°’ì„ append
            #print( i ,"new:",new)
            
    return new.T    
```

# 2) PCAë¥¼ êµ¬í˜„í•´ë³´ê¸°

ìœ„ì˜ ê³¼ì •ì„ í•˜ë‚˜ì˜ í•¨ìˆ˜ë¡œ í‘œí˜„í•´ì£¼ì„¸ìš”!

1)ì—ì„œ ì§„í–‰í•œ ê³¼ì •ì„ ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜¤ë©´ ë©ë‹ˆë‹¤..! , 

ìœ„ì—ì„œ ë§Œë“  í•¨ìˆ˜(new_coordinate)ë„ ê°€ì ¸ì˜¤ì…”ì„œ ì‚¬ìš©í•´ë„ ì¢‹ì•„ìš”!

+í•˜ì§€ë§Œ ì´ í•¨ìˆ˜ëŠ” numberì´ë¼ëŠ” ì¸ìë¥¼ ë°›ì•„ì„œ pcaì—ì„œ ì£¼ì„±ë¶„ ê°œìˆ˜ë¥¼ ê³ ë ¤í•˜ë„ë¡ í•©ë‹ˆë‹¤!

ê³ ìœ ê°’ì´ í° ì¶• ìˆœì„œë¡œ ì •ë ¬í•˜ëŠ” ê³¼ì • + ì£¼ì„±ë¶„ ê°œìˆ˜ numberì„ ê³ ë ¤í•´ì•¼í•œë‹¤ëŠ” ì ì„ ìœ ì˜í•´ì„œ í•¨ìˆ˜ë¥¼ ì •ì˜í•´ì£¼ì„¸ìš”

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
  # Scaling
    scaler = StandardScaler()
    X_std = scaler.fit_transform(X)
    features = X_std.T
    
  # Covariance
    cov_matrix = np.cov(features)

  # eigenvalues, eigenvectors
    eigenvalues = lin.eig(cov_matrix)[0]
    eigenvectors = lin.eig(cov_matrix)[1]    

  #np.dot(np.dot(eigenvectors,mat),eigenvectors.T)

  # ê³ ìœ ê°’ ì •ë ¬ ë° ê³ ìœ ê°’ í¬ê¸°ì— ë”°ë¼ ê³ ìœ ë²¡í„°ë„ ì •ë ¬
    idx = eigenvalues.argsort()[::-1] # ê³ ìœ ê°’ì„ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œ ìˆœì„œ
    eigenvalues = eigenvalues[idx] # ì•ì„œ êµ¬í•œ ìˆœì„œëŒ€ë¡œ ê³ ìœ ê°’ì„ ì •ë ¬
    eigenvectors = eigenvectors.T[idx].T # ì•ì„œ êµ¬í•œ ìˆœì„œëŒ€ë¡œ ê³ ìœ ë²¡í„°ë„ ì •ë ¬
    

  #def new_coordinates(X,eigenvectors): ì•ì„œ êµ¬í•œ ì •ì˜ í™œìš©
    for i in range(min(number,eigenvectors.shape[0])):
    
        if i == 0:
            new = [np.dot(X,eigenvectors[:,i])] 
            # Xì™€ eigenvectorì„ ë‚´ì í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ ìƒì„±
            # print( i ,"new:",new)
        else: 
            new = np.concatenate((new, np.dot(X, eigenvectors.T[i]).reshape(1,-1)),axis=0) # reshapeë¥¼ ì‚¬ìš©í•´ ë³´ì•˜ë‹¤.    
            # concatencateë¥¼ ì´ìš©í•˜ì—¬ new list ì— Xì™€ i(1,2)ë¥¼ ë‚´ì í•œ ê°’ì„ append
            #print( i ,"new:",new)


    return new.T #ìƒˆë¡œìš´ ì¶•ìœ¼ë¡œ ë³€í™˜ë˜ì–´ ë‚˜íƒ€ë‚œ ë°ì´í„° return
```

```python
MYPCA(X,3)

# ìƒˆë¡œìš´ ì¶•ìœ¼ë¡œ ì˜ ë³€í™˜ë˜ì–´ì„œ ë‚˜íƒ€ë‚˜ë‚˜ìš”? ë„µ
# ìœ„ì—ì„œ í–ˆë˜ PCAë‘ì€ ì°¨ì´ê°€ ìˆì„ ìˆ˜ ìˆì–´ìš” 
# ì™œëƒí•˜ë©´ ìœ„ì—ì„œëŠ” ê³ ìœ ê°’ì´ í° ì¶• ìˆœì„œë¡œ ì •ë ¬ì„ ì•ˆí–ˆì—ˆê±°ë“ ìš”
```
